## Import Statements

```python
import pandas as pd
import datetime
from IPython.display import display
```

## Load and Inspect Spotify data

```python
spotify = pd.read_csv('spotifydata.csv')
spotify = spotify.rename(index=str, columns={"name":"song_title"})

display(spotify)
```

## Load and Inspect Grammy Winner data

```python
winners = pd.read_csv('grammywinners.csv')
winners = winners.rename(index=str, columns={"nominee":"song_title"})

display(winners.dtypes)
```

```python
start_date = "1999"
end_date = "2019"
```

## Merge Winners with Spotify data

```python
df = pd.merge(winners, spotify, on='song_title', how='left')

display(df)
df.to_csv(r'C:\Users\junea\Projects\CapstoneTwo\winner_spotify_join.csv',index=False)
```
